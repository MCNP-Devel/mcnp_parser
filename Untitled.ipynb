```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import mcnp_parser as mcnp
```

```python
spectrum_7 = mcnp.read_output("./mcnp_file/700k.o")
spectrum_6 = mcnp.read_output("./mcnp_file/600k.o")
```

```python
for i in range(len(spectrum_6)):
    if(i %2 == 0):
        plt.figure()
        plt.semilogx()
        plt.grid()
        x =  [val[0] for val in spectrum_6[i]]
        y_6 =  [val[1] for val in spectrum_6[i]]
        y_7 =  [val[1] for val in spectrum_7[i]]
        y = [ (y_7[i] - y_6[i])/(y_7[i]) for i in range(len(y_7)) ]
        
        y_err_6 =  [val[2] for val in spectrum_6[i]]
        y_err_7 =  [val[2] for val in spectrum_7[i]]
        y_err = [ (y_err_7[i] + y_err_6[i]) for i in range(len(y_7)) ]

        
        plt.errorbar(x, y, xerr=0.0, yerr=y_err, label =str(i), linewidth=.3)
        name = "myspectrum_"  + str(i) + ".png"
        plt.savefig(name, dpi=500)
```
